```python
import seaborn as sns
#stat = 'probability', bins = 'sqrt', kde = True
# sns.histplot(df['LWdown'], bins = 'sqrt', kde = True)

#widget for interactive plot 
import ipywidgets as widgets
from ipywidgets import interact, interact_manual
import matplotlib.pyplot as plt
import numpy as np

class PlotHist:
    def __init__(self, df):
        self.df = df
    def plot_hist(self, df, stat, bins, kde):
        sns.histplot(df[stat], bins = bins, kde = kde)
        plt.show()

    def plot_hist_interact(self, df):
        #only works for numerical columns
        stat = widgets.Dropdown(
            options = df.columns,
            value = df.columns[0],
            description = 'stat',
            disabled = False,
        )
        bins = widgets.Dropdown(
            options = ['auto', 'sqrt', 'sturges', 'rice', 'scott', 'fd', 'doane', 'harris'],
            value = 'auto',
            description = 'bins',
            disabled = False,
        )
        kde = widgets.Dropdown(
            options = [True, False],
            value = True,
            description = 'kde',
            disabled = False,
        )

        interact(self.plot_hist, df = widgets.fixed(df), stat = stat, bins = bins, kde = kde)


    #INTERACTIVE WITH sns.widget
    def with_sns(self, df):
        stat = widgets.Dropdown(
            options = df.columns,
            value = df.columns[0],
            description = 'stat',
            disabled = False,
        )
        bins = widgets.Dropdown(
            options = ['auto', 'sqrt', 'sturges', 'rice', 'scott', 'fd', 'doane', 'harris'],
            value = 'auto',
            description = 'bins',
            disabled = False,
        )
        kde = widgets.Dropdown(
            options = [True, False],
            value = True,
            description = 'kde',
            disabled = False,
        )

        @interact(stat = stat, bins = bins, kde = kde)
        def plot_hist(stat, bins, kde):
            sns.histplot(df[stat], bins = bins, kde = kde)
            plt.show()

    #INTERACTIVE WITH matplotlib.pyplot
    def with_matplotlib(self, df):
        stat = widgets.Dropdown(
            options = df.columns,
            value = df.columns[0],
            description = 'stat',
            disabled = False,
        )
        bins = widgets.Dropdown(
            options = ['auto', 'sqrt', 'sturges', 'rice', 'scott', 'fd', 'doane', 'harris'],
            value = 'auto',
            description = 'bins',
            disabled = False,
        )
        kde = widgets.Dropdown(
            options = [True, False],
            value = True,
            description = 'kde',
            disabled = False,
        )

        #combine two inputs into one
        def plot_hist(stat, bins, kde):
            sns.histplot(df[stat], bins = bins, kde = kde)
            plt.show()


# df = df.drop(columns = ['_sa_instance_state'])
plot_hist = PlotHist(df)
plot_hist.with_sns(df)
```
